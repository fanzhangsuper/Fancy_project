# Process data

```python
with open("data/twitter-swisscom/twex.tsv", 'r') as h:
    lines = h.readlines()
```

```python
to_be_processed_lines = [i for k, i in enumerate(lines) if i.endswith('\\\n') ]
print(len(to_be_processed_lines))
```

```python
def remove_line_breaks(list_of_strings):
    count = 0
    for i in range(len(list_of_strings)):
        list_of_strings[i] = list_of_strings[i].replace('\\\t', '')
        if list_of_strings[i].endswith('\\\n'):
            list_of_strings[i] = list_of_strings[i][:-2]
            
remove_line_breaks(lines)
```
