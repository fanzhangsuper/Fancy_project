## The Aim of Graphical Disply

    1. show the data
    2. induce the viewer to think about the substance rather than about moethodology, graphical design or sth else
    3. avoid distorting data
    4. encourage the ey to compare different pieces of data
    5. reveal the data at several levels of detail, from a broad overview to the fine structure
    6. serve a reasonably clear purpose: description, exploration, tabulation, or decoartion
    7. be closely integrated with the statistical and verbal descriptions of a data set.

Graphics should be reserved for the richer, more complex more difficult statistical material. If we can summarize data in one or two numbers, we should not use graphs.

## Maps

Advantage

    1. direct attention toward exploring the substantive content of the data
    2. less computation
    3. extensive data
    
Disadvantage
    
    1. wrongly equate the visual importance of each county with its geographic area rather than with the number of people living in the county.
    2. maps are founded on a suspect data source. For example, regional clustering may reflect varying diagnostic customs.

# Data-Ink

Data iink is the non-erasable core of a graphic, the non-redundant ink arranged in response to variation in the number s represented.
$$
\text{data ink ratio} = \frac{\text{data-ink}}{\text{total ink used to print the graphic}}
$$

The fundamental pricnciple of good statistical graphics:

    Above all else show the data.
    Maximize the data ink ratio.
    Erase non-data-ink
    Erase redundant data-ink
    Revise and edit.
    
But there are exceptions, like we want to concatenate timetable of trains.

Bad examples are like: 

    grid sea overwhelms the numbers.

## Data Density

High density graphics help us to compare parts of the data by displaying much information within the view of the eye: we look at one page at a time and the more on the page, the more effective and comparative our eye can be. The principle is:

    Maximize data density and the size of the data matrix, within reason.
    
Special care: as the volumne of data increases, data measures must shrink (smaller dots for scatters, thinner lines for busy time-series).

The way to increase data density other than by enlarging the data matrix is to reduce the area of a graphic. The **Shrink Principle** has wide application:
    
    Graphics can be shrunk way down.

Data reduction: averaging, clustering, smoothing

#### Small Multiples

Well-designed small multiples are:

- inevitably comparative
- deftly multivariate
- shrunken, high-density graphics
- usually based on a large data matrix
- drawn almost entirely with data-ink
- efficient in interpretation
- often narrative in content, showing shifts in the relationship between interaction or multiplicative effects.

Small multiples reflect much of the thoery of data graphics:

    For non-data-ink, less is more
    For data-ink, less is bore

## Aesthetics and Technique in Data Graphical Design

    Graphical elegance is often found in simplicity of design and complexity of data.

Here are some guides for enhancing the visual quality of the more routine, workaday designs. Attractive displays of information

- have a properly chosen format and design
- use words, numbers, and drawing together
- reflect ad balance, a proportion, a sense of relevant scale
- display an accessible complexity of detail
- often have a narrative quality, a story to tell about the data
- are drawn in a professional manner, with the technical details of production done with care
- avoid content-free decoration, including charjunk.
