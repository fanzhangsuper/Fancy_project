This notebook is used to process data. From the original .tsv file, we try to extract the useful information and process the part of data for future use.

Based on aim of the project, we try to analyse the mobility of switzerland and some neighbor countries by twitter data. So from twitter we get, the useful information is the placeinformation, post time, text content.

Considered of the number of data, we divided the data into 4 part. The file shows the 1/4 data's process and we store the data processing result as .pickle file. The rest of data was processed by the seem method.

First, we delete the part of missing data and store the data into dataframe with column 'userId', 'createdAt','placeId','placeLatitude', 'placeLongitude','followersCount','friendsCount','statusesCount'.

Then we try to extract year, month, day, weekday, hour informtion from column 'createAt'. For this, we convert the data in colume 'createAt' into type of datatime, and add columns 'month', 'weekday', 'day', 'hour' respectively.
We store the time data as a file called 'twitter_create_time.pickle'.

Third, we store the place information: 'placeId', 'placeLatitude', 'placeLongitude' as file called 'geo_info.pickle'.

Another important process of data is about geography information. From the placeid, placeLatitude', 'placeLongitude', we can get information through the library 'geopy'. We extract the country_code, state respective to a specific placeid. Then we store the corresponding relation of 'placeid','country_code', 'state' as file called 'pId_geo_df'.

Besides, we store the users reletive information 'userId', 'followersCount', 'friendsCount' as file called 'user_info'.

Finally, we count number of twiiter on different time and different place.

```python
from geopy.geocoders import Nominatim
from geopy.distance import vincenty

# >>> from geopy.geocoders import Nominatim
# >>> geolocator = Nominatim()
# >>> location = geolocator.geocode("175 5th Avenue NYC")
# >>> print(location.address)
# Flatiron Building, 175, 5th Avenue, Flatiron, New York, NYC, New York, ...
# >>> print((location.latitude, location.longitude))
# (40.7410861, -73.9896297241625)
# >>> print(location.raw)
# {'place_id': '9167009604', 'type': 'attraction', ...}

def geo(loc):
    """ Input a string of loc"""
    geolocator = Nominatim()
    location = geolocator.geocode(loc)
    return location.longitude, location.latitude

def rev_geo(lon, lat):
    """Use geopy to do reverse geocoding
    
    Becarefule of the output format as geopy has multiple sources and can have many format.
    """
    geolocator = Nominatim()
    location = geolocator.reverse(str(lat) + ", " + str(lon))
    return location.raw

def distance(lon1, lat1, lon2, lat2):    
    newport_ri = (lon1, lat1)
    cleveland_oh = (lon1, lat2)
    return vincenty(newport_ri, cleveland_oh).kilometers
```

```python
def get_place_name_of_placeId_from_df(df, pause_secs=1):
    """The input dataframe is """
    import time
    if (df.index.name != 'placeId' or any(df.index != df.index.unique())):
        raise ValueError("Input dataframe should use uniuqe 'placeId' as index")

    for col in ['placeLongitude', 'placeLatitude']:
        if (col not in df.columns):
            raise ValueError("Input dataframe should have cols 'placeLongitude', 'placeLatitude'")

    pId_geo_map = []
    t = time.time()
    try:
        for i, pId in enumerate(df.index):
            x = df.loc[pId]
            raw_geo_data = rev_geo(lon=x['placeLongitude'], lat=x['placeLatitude'])
            pId_geo_map.append((pId, raw_geo_data))
            time.sleep(pause_secs)
            print(i, time.time() - t)
            t = time.time()
    except Exception as e:
        print("Expected Error " + e)
        return pId_geo_map

    return pId_geo_map
```
